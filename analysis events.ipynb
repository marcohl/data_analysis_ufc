```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_events = pd.read_csv('ufc_event_data.csv')
df_events.head(20)
```

```python
#Cleaning of the dataframe
# Replace the "-----" values with NaN in KD
df_events['KD'] = df_events['KD'].replace('-----', np.nan)

# Drop the rows containing NaN values
df_events.dropna(subset=['KD'], inplace=True)

# Split the KD column into two separate columns
df_events[['KD_1', 'KD_2']] = df_events['KD'].str.split('-', expand=True)

# Drop the original KD column
df_events.drop('KD', axis=1, inplace=True)
```

```python
# Replace the "-----" values with NaN in TD
df_events['TD'] = df_events['TD'].replace('-----', np.nan)

# Drop the rows containing NaN values
df_events.dropna(subset=['TD'], inplace=True)

# Split the TD column into two separate columns
df_events[['TD_1', 'TD_2']] = df_events['TD'].str.split('-', expand=True)

# Drop the original TD column
df_events.drop('TD', axis=1, inplace=True)
```

```python
# Replace the "-----" values with NaN in Sub
df_events['Sub'] = df_events['Sub'].replace('-----', np.nan)

# Drop the rows containing NaN values
df_events.dropna(subset=['Sub'], inplace=True)

# Split the TD column into two separate columns
df_events[['Sub_1', 'Sub_2']] = df_events['Sub'].str.split('-', expand=True)

# Drop the original Sub column
df_events.drop('Sub', axis=1, inplace=True)
```

```python
# Replace the "-----" values with NaN in Strikes
df_events['Strikes'] = df_events['Strikes'].replace('-----', np.nan)

# Drop the rows containing NaN values
df_events.dropna(subset=['Strikes'], inplace=True)

# Split the Strikes column into two separate columns
df_events[['Strikes_1', 'Strikes_2']] = df_events['Strikes'].str.split('-', expand=True)

# Drop the original Strikes column
df_events.drop('Strikes', axis=1, inplace=True)
```

```python
df_events['Round'].info()
```

```python
# Split the "time" column into two separate columns for minutes and seconds
df_events[['Minutes', 'Seconds']] = df_events['Time'].str.split(':', expand=True)

# Convert the "minutes" and "seconds" columns to integer type
df_events['Minutes'] = df_events['Minutes'].astype(int)
df_events['Seconds'] = df_events['Seconds'].astype(int)

# Calculate the total time in seconds
df_events['total_time_seconds'] = (df_events['Minutes'] * 60 + df_events['Seconds'])* df_events['Round']

# Drop the original "time" column
df_events.drop(['Time'], axis=1, inplace=True)
```

```python
#Convert the object type columns to int type
df_events['KD_1'] = df_events['KD_1'].astype(int)
df_events['KD_2'] = df_events['KD_2'].astype(int)
df_events['TD_1'] = df_events['TD_1'].astype(int)
df_events['TD_2'] = df_events['TD_2'].astype(int)
df_events['Sub_1'] = df_events['Sub_1'].astype(int)
df_events['Sub_2'] = df_events['Sub_2'].astype(int)
df_events['Strikes_1'] = df_events['Strikes_1'].astype(int)
df_events['Strikes_2'] = df_events['Strikes_2'].astype(int)
df_events.info()
```

```python
# Concatenate Fighter1 and Fighter2 columns to create a single column
df_events_copy['Fighter'] = df_events_copy['Fighter1'].fillna('') + ' / ' + df_events_copy['Fighter2'].fillna('')

# Create a new column to store the total number of submissions
df_events_copy['Total_Submissions'] = df_events_copy['Sub_1'].fillna(0) + df_events_copy['Sub_2'].fillna(0)

# Group the DataFrame by Fighter and calculate the sum of Total_Submissions
submission_ranking = df_events_copy.groupby('Fighter')['Total_Submissions'].sum().reset_index()

# Sort the DataFrame in descending order based on the total number of submissions
submission_ranking = submission_ranking.sort_values(by='Total_Submissions', ascending=False)
submission_ranking.head()

```

```python
# Create a copy of the original DataFrame
df_copy = df_events_copy.copy()

# Create a new DataFrame to store the total number of submissions for each fighter
submission_totals = pd.DataFrame(columns=['Fighter', 'Total_Submissions'])

# Iterate over each unique fighter in the DataFrame
for fighter in pd.unique(df_copy[['Fighter1', 'Fighter2']].values.ravel()):
    # Filter rows where the current fighter appears in either Fighter1 or Fighter2 column
    mask = (df_copy['Fighter1'] == fighter) | (df_copy['Fighter2'] == fighter)
    
    # Calculate the total number of submissions for the current fighter
    total_submissions = df_copy.loc[mask, 'Sub_1'].sum() + df_copy.loc[mask, 'Sub_2'].sum()
    
    # Append the fighter and total submissions to the submission_totals DataFrame
    submission_totals = pd.concat([submission_totals, pd.DataFrame({'Fighter': [fighter], 'Total_Submissions': [total_submissions]})])

# Sort the DataFrame in descending order based on the total number of submissions
submission_totals = submission_totals.sort_values(by='Total_Submissions', ascending=False)

# Print the ranking of fighters with the most total submissions
submission_totals.head(10)

```

```python
#Check if there are fighters that have fought in 2 or more weight categories
# Get a list of all unique fighters in the DataFrame
fighters = df_events_copy["Fighter1"].append(df_events_copy["Fighter2"]).unique()

# Create a dictionary to store the number of weight classes each fighter has fought in
weight_classes = {}

# Check how many weight classes each fighter has fought in
for fighter in fighters:
    weight_classes_set = set(df_events_copy.loc[(df_events_copy["Fighter1"] == fighter) | (df_events_copy["Fighter2"] == fighter), "Weight Class"])
    weight_classes[fighter] = len(weight_classes_set)

# Sort the dictionary by value (number of weight classes) in descending order
weight_classes_sorted = sorted(weight_classes.items(), key=lambda x: x[1], reverse=True)

# Print the results
for fighter, num_weight_classes in weight_classes_sorted:
    if num_weight_classes >= 2:
        print(fighter + " has fought in " + str(num_weight_classes) + " weight classes.")
```

```python
# Calcular el promedio de Strikes_1 y Strikes_2
df_events_copy['Strikes_1'] = df_events_copy['Strikes_1'].astype(float)
df_events_copy['Strikes_2'] = df_events_copy['Strikes_2'].astype(float)
average_strikes_1 = df_events_copy['Strikes_1'].mean()
average_strikes_2 = df_events_copy['Strikes_2'].mean()

# Crear un gráfico de dispersión con líneas de regresión
sns.regplot(x='Strikes_1', y='Strikes_2', data=df_events_copy)

# Establecer el título y etiquetas del gráfico
plt.title('Promedio de Strikes_1 vs Strikes_2')
plt.xlabel('Strikes_1')
plt.ylabel('Strikes_2')

# Agregar texto con el promedio de Strikes_1 y Strikes_2
plt.text(280, 70, f'Promedio Strikes_1: {average_strikes_1:.2f}')
plt.text(280, 40, f'Promedio Strikes_2: {average_strikes_2:.2f}')

# Mostrar el gráfico
plt.grid()
plt.show()
```

```python
# Ordenar las categorías de Weight Class
weight_class_order = ['Women\'s Bantamweight', 'Women\'s Strawweight', 'Bantamweight', 'Flyweight',
                      'Featherweight', 'Lightweight', 'Welterweight', 'Middleweight',
                      'Light Heavyweight', 'Heavyweight']

# Crear una figura de tipo "box" con los boxplots de Strikes_1 y Strikes_2
sns.catplot(x='Weight Class', y='value', hue='variable', data=df_events_copy.melt(id_vars='Weight Class', value_vars=['Strikes_1', 'Strikes_2']), kind='box', order=weight_class_order, aspect=2)

# Configurar el título y las etiquetas de los ejes
plt.title('Comparison - Strikes_1 and Strikes_2 per Weight Class')
plt.xlabel('Weight Class')
plt.ylabel('Number of strikes')

# Rotar las etiquetas del eje x
plt.xticks(rotation=90)

# Mostrar el gráfico
plt.grid()
plt.show()
```

```python
#Fighter with most wins per weight category
# Group the DataFrame by "Weight Class" and "Fighter1"
grouped_by_weight_class = df_events_copy.groupby(['Weight Class', 'Fighter1'])

# Count the number of wins for each fighter in each weight class
wins_count = grouped_by_weight_class.apply(lambda x: (x['Result'] == x['Fighter1']).sum())

# Find the fighter with the most wins for each weight class
most_wins_per_weight_class = wins_count.groupby(level='Weight Class').idxmax().apply(lambda x: x[1])

# Get the corresponding number of wins for each fighter
most_wins_count_per_weight_class = wins_count.groupby(level='Weight Class').max()

# Create a DataFrame to store the results
results_df = pd.DataFrame({'Fighter': most_wins_per_weight_class, 'Wins': most_wins_count_per_weight_class})

# Print the results
print(results_df)
```
